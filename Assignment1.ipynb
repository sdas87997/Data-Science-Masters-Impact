```python

------------------------------------------------------------------------------------------------------------------------------------------------


Q1.  What are the characteristics of the tuples? Is tuple immutable?

Solution :-  Tuples have the given characteristics:-
•	They are indexed
•	Tuples are ordered
•	These are immutable
•	They can contain duplicate items
Tuples are immutable, This means that we cannot change their contents after they are created.
Example:- Let us create a tuple a with values 1,2,3,4,5,6and 6
a = (1,2,3,4,5,6,6)
a[0] ,here at 0 index 1 is present , so we get 1 as output
but if we want to assign any value to a[0] , then we get error because this assignment is not possible due to immutability
 
---------------------------------------------------------------------------
TypeError                                 Traceback (most recent call last)
Cell In[27], line 1
----> 1 a[0] = 12

TypeError: 'tuple' object does not support item assignment


----------------------------------------------------------------------------------------------------------------------------------------------
Q2. What are the two tuple methods in python? Give an example of each method. Give a reason why
tuples have only two in-built methods as compared to Lists.

Solution:-  There are only two tuple methods count() and index() that a tuple object can call.
The count function returns the count of the element in the tuple
The index function returns the index of the element in the tuple

For example Let us create a tuple a with values 1,2,3,4,5,6and 6
a = (1,2,3,4,5,6,6)
 

a.index(3) – This finds the index of the value  3 in the tuple and 3 is present in the 2nd index so output is 2

a.count(3) – this finds how many times 3 is present inside the tuple, here 3 is present once so output is 1

Lists have several built-in methods. A unexpected change or error is more likely to occur in a list. In a tuple, changes and errors don't usually occur because of immutability.


------------------------------------------------------------------------------------------------------------------------------------------------
Q3. Which collection datatypes in python do not allow duplicate items? Write a code using a set to remove
duplicates from the given list.
List = [1, 1, 1, 2, 1, 3, 1, 4, 2, 1, 2, 2, 2, 3, 2, 4, 3, 1, 3, 2, 3, 3, 3, 4, 4, 1, 4, 2, 4, 3, 4, 4]

solution:- 
            List1 = [1, 1, 1, 2, 1, 3, 1, 4, 2, 1, 2, 2, 2, 3, 2, 4, 3, 1, 3, 2, 3, 3, 3, 4, 4, 1, 4, 2, 4, 3, 4, 4]
    
       print("List1---",List1)
          setc = set(List1)
            print("After converting to set------",setc)
            
output:- 
           List1--- [1, 1, 1, 2, 1, 3, 1, 4, 2, 1, 2, 2, 2, 3, 2, 4, 3, 1, 3, 2, 3, 3, 3, 4, 4, 1, 4, 2, 4, 3, 4, 4]
           After converting to set------ {1, 2, 3, 4}
------------------------------------------------------------------------------------------------------------------------------------------------




Q4. Explain the difference between the union() and update() methods for a set. Give an example of
each method.

solution:- The update() adds all missing elements to the set on which it is called .The set. union() creates a new set and Consequently, the return value of set.
                       s.union(t)  s | t   new set with elements from both s and t
                       s.update(t) s |= t  return set s with elements added from t
    example:-
               s1={1,2,3,4}
           print("Before Update",s1)
            s1.update({5})
            print("After Update the original set s1 gets modified",s1)
            s8 = {7,8}.union(s1)
            print("After union s1 no impact on s1, but union generates a new set , so s1 is :-",s1)
            print("After union s8 new set born with all elements, so s8 is",s8)
    output:-
                 Before Update {1, 2, 3, 4}
                 After Update the original set s1 gets modified {1, 2, 3, 4, 5}
                 After union s1 no impact on s1, but union generates a new set , so s1 is :- {1, 2, 3, 4, 5}
                 After union s8 new set born with all elements, so s8 is {1, 2, 3, 4, 5, 7, 8}

-----------------------------------------------------------------------------------------------------------------------------------------------
Q5. What is a dictionary? Give an example. Also, state whether a dictionary is ordered or unordered.
Dictionaries are collections in pythons which are used to store data values in key:value  pairs.
A dictionary is a collection which is ordered, changeable and do not allow duplicates.
When we say that dictionaries are ordered, it means that the items have a defined order, and that order will not change.
Unordered means that the items does not have a defined order, we cannot refer to an item by using an index.
But in dictionary, the keys function as indexes and we can easily refer them.
Dict1 = {
  "brand": "Ford",
  "model": "Mustang",
  "year": 1964,
  "year": 2020
}
print(Dict1)

 output:- {'brand': 'Ford', 'model': 'Mustang', 'year': 2020}
    
    
-- --------------------------------------------------------------------------------------------------------------------------------------   
    Q6. Can we create a nested dictionary? If so, please give an example by creating a simple one-level
nested dictionary.

Solution:- Yes we can create a nested dictionary.
Example of a simple one-level nested dictionary. is

Dict1 = {
  "brand": "Ford",
  "modeldetailss": {"name":"Mustang","milegae":"20km/hr"},
  "year": 1964,
  "year": 2020
}
print(Dict1)
in output:-    
{'brand': 'Ford', 'modeldetailss': {'name': 'Mustang', 'milegae': '20km/hr'}, 'year': 2020}

here 'modeldetailss': {'name': 'Mustang', 'milegae': '20km/hr'} is a nested dictionary.
-----------------------------------------------------------------------------------------------------------------------------
Q7. Using setdefault() method, create key named topics in the given dictionary and also add the value of
the key as this list ['Python', 'Machine Learning’, 'Deep Learning']
                      
solution:-    dict1={}
              dict1.setdefault('language' , 'Python')
              dict1.setdefault('course' , 'Data Science Masters')
              print(dict1)
output:-
             {'language': 'Python', 'course': 'Data Science Masters'}         

-----------------------------------------------------------------------------------------------------------------------------

Q8. What are the three view objects in dictionaries? Use the three in-built methods in python to display  these three view objects for the given dictionary.
dict1 = {'Sport': 'Cricket' , 'Teams': ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']}
Ans:- The three view objects in dictionaries in Python are Key, Value and Items
            The  three in-built methods are :
1>	.keys() :- This method is used to find out the keys of the key value pairs present in the dictionary.
dict1 = {"Sport": "Cricket", "Teams": ["India", "Australia", "England", "South Africa", "Sri Lanka", "New Zealand"]}
dict1.keys()  ---- In output we get  
 dict_keys(['Sport', 'Teams'])

2>	.values() :- This method is used to find out the Values of the key value pairs present in the dictionary.
dict1 = {"Sport": "Cricket", "Teams": ["India", "Australia", "England", "South Africa", "Sri Lanka", "New Zealand"]}
dict1.values()  ---- In output we get  
dict_values(['Cricket', ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand']])
 

3>	 .items():- This method is used to find out the keys along with Values of the key value pairs present in the dictionary.It put the key value pair in tuples.
dict1 = {"Sport": "Cricket", "Teams": ["India", "Australia", "England", "South Africa", "Sri Lanka", "New Zealand"]}
dict1.items()  ---- In output we get  
 dict_items([('Sport', 'Cricket'), ('Teams', ['India', 'Australia', 'England', 'South Africa', 'Sri Lanka', 'New Zealand'])])



--------------------end of Assignment--------------------------------------------------------------------------------






```
