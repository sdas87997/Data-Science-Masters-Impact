```python
import numpy as np
from scipy.special import comb
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1
red_balls = 10
blue_balls =15
total_balls =red_balls + blue_balls
prob_of_red_ball = red_balls/total_balls
prob_of_blue_ball = blue_balls/total_balls
print(red_balls,blue_balls,total_balls,prob_of_red_ball,prob_of_blue_ball)
no_ball_drwn = 12
probality_arr = []
for ctr in range (1,11):
    prob =  comb(no_ball_drwn,ctr) * (prob_of_red_ball**ctr) *  (prob_of_blue_ball** (no_ball_drwn - ctr))
    probality_arr.append(prob)  
print(probality_arr)
plt.plot(range(1,11),probality_arr)
plt.bar(range(1,11),probality_arr)

#expected mean
expected_red = sum(np.array(probality_arr)*list(range(1,11)))
expected_blue = sum(np.array(probality_arr)*list(range(11,1,-1)))
print("expected_red",expected_red)
print("expected_blue",expected_blue)

#expected variance
variance_red_arr = (list(range(1,11) - expected_red))
x1 = [j**2 for j in variance_red_arr]
variance_red = sum(x1 * np.array(probality_arr))

variance_blue_arr = (list(range(11,1,-1) - expected_blue))
x1 = [j**2 for j in variance_blue_arr]
variance_blue = sum(x1 * np.array(probality_arr))
print("variance_red",variance_red)
print("variance_blue",variance_blue)
                   

```

```python
import numpy as np
from scipy.special import comb
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1
red_balls = 10
blue_balls =15
total_balls =red_balls + blue_balls
prob_of_red_ball = red_balls/total_balls
prob_of_blue_ball = blue_balls/total_balls
print(red_balls,blue_balls,total_balls,prob_of_red_ball,prob_of_blue_ball)
no_ball_drwn = 12
probality_arr = []
for ctr in range (1,11):
    prob =  comb(no_ball_drwn,ctr) * (prob_of_red_ball**ctr) *  (prob_of_blue_ball** (no_ball_drwn - ctr))
    probality_arr.append(prob)  
print(probality_arr)
cum_probs =  np.cumsum(probality_arr)
plt1.plot(range(1,11),cum_probs)
plt1.bar(range(1,11),cum_probs)
```
