Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

In Bokeh, glyphs are visual markers or symbols that represent data points on a plot. 
They can be used to create scatter plots, line plots, bar charts, and various other types of visualizations.
To add glyphs to a Bokeh plot, you typically use the figure
object's glyph methods such as circle(), line(), square(), etc., to specify the type of glyph you want to use and its properties.


```python
#code of previous explanation
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure , output_file, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
output_file("plot.html")
p = figure(title="My Bokeh Plot")
p.circle(x, y, size=10, color="navy", alpha=0.5)
show(p)
```

Here we use the circle() method to add a  glyph.

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()

from bokeh.plotting import figure , output_file, show

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]
output_file("plot.html")
p = figure(title="My Bokeh Plot")
p.xaxis.axis_label = "xaxis"
p.yaxis.axis_label = "yaxis"
p.scatter(x, y, size=10, fill_color="red", legend_label="Datapoints")
show(p)
```

real time?

```python
A Bokeh server is a Python process that allows you to create and serve interactive Bokeh plots and applications.
It provides a way to create dynamic and interactive visualizations that can be updated in real-time based on user interactions or data changes.
To use the Bokeh server and create interactive plots that can be updated in real-time, you need to follow these steps:

Step 1: Import the necessary modules and functions from Bokeh.


from bokeh.plotting import figure, curdoc
from bokeh.models import ColumnDataSource
from bokeh.layouts import row
from bokeh.io import output_notebook, show
Step 2: Prepare the initial data and create a ColumnDataSource object to hold the data that will be updated.


# Prepare initial data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(x=x, y=y))
Step 3: Create the interactive plot(s) using the figure function.


# Create a figure
p = figure(title="Interactive Plot", x_axis_label='X', y_axis_label='Y', plot_width=400, plot_height=400)

# Add glyphs using the ColumnDataSource
p.circle('x', 'y', size=10, color="navy", alpha=0.5, source=source)
Step 4: Define callback functions that will update the plot when triggered by user interactions or data changes.


# Define a callback function
def update_data():
    new_x = [1, 2, 3, 4, 5]
    new_y = [6, 7, 2, 4, 5]
    source.data = dict(x=new_x, y=new_y)

# Call the update_data function when a button is clicked
button = Button(label="Update Data")
button.on_click(update_data)
Step 5: Create a layout for the plot(s) and any other widgets or controls.


# Create a layout
layout = row(p, button)
Step 6: Create the Bokeh document and add the layout to it.


# Create the Bokeh document
curdoc().add_root(layout)
Step 7: Run the Bokeh server.


# Run the Bokeh server
output_notebook()  # Optional for Jupyter notebooks
show(layout)       # Optional for non-server usage
By running the Bokeh server, you can now view the interactive plot and any associated controls or widgets in your web browser. 
The plot will update in real-time based on the interactions defined in the callback function(s), such as clicking a button to update the data.





```

Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?


To embed a Bokeh plot into a web page or dashboard using Flask or Django, you need to follow these general steps:

Create a Bokeh plot as usual using the Bokeh library. This involves importing the necessary modules, creating the plot, and configuring its properties.

Convert the Bokeh plot to HTML using the bokeh.embed module. This module provides functions to generate the necessary HTML and JavaScript code to embed the plot into a web page.

Integrate the Bokeh plot into your Flask or Django application by following the appropriate framework-specific steps:

For Flask:

Import the necessary modules from Flask.
Define a route (URL endpoint) that will render the web page containing the Bokeh plot.
In the route function, call the bokeh.embed.components function to generate the HTML and JavaScript code for the plot.
Pass the generated code to the HTML template and return the rendered template as the response.
