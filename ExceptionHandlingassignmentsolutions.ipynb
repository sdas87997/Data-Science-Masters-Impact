```python
#Q1. Explain why we have to use the Exception class while creating a Custom Exception.
#Note: Here Exception class refers to the base class for all the exceptions.
When creating a custom exception in Python, it is recommended to inherit from the built-in Exception class or one of its subclasses.
Here are the reasons why using the Exception class is preferred:
Standardization: Inheriting from Exception or its subclasses helps maintain consistency and adheres to the established Python exception hierarchy.
This hierarchy allows for organized exception handling and clear understanding of exception types.
Exception Handling: Python provides a powerful exception handling mechanism that allows you to catch and handle exceptions in a structured way.
By inheriting from Exception, the custom exception will be compatible with the existing exception handling infrastructure. 
so we can catch your custom exception using a specific except block or handle it along with other exceptions in a more general manner.
Robustness: The Exception class provides essential functionality that is often needed when dealing with exceptions. 
It includes attributes like args (arguments passed to the exception), message (the exception message), 
and methods like __str__() (to generate a string representation of the exception). By inheriting from Exception, 
the custom exception will inherit these capabilities, making it more robust and easier to work with.
```

```python
#Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class in Python is a base class for exceptions that occur when a key or index used to look up a value in a 
mapping or sequence is not found. It serves as a superclass for several specific lookup-related exceptions, 
providing a consistent way to handle lookup failures.

The LookupError class is useful because it allows you to catch and handle lookup-related exceptions in a more general manner, 
without needing to handle each specific exception separately. By catching LookupError, you can handle any exception 
that is a subclass of LookupError, which includes the specific lookup-related exceptions defined in Python.

-------KeyError--------------

#Exception
try:
    dicr1 = {"name":"supriyo", "age": 21}
    print(dicr1["name"])
    print(dicr1["gender"])
except KeyError as e:
    print("exception in code",e)
else:
    print("Hurray try block executed without any exception!!!!")
finally:
    print("I don't care , mai rookne nahi wala,bhar mai jay duniya and exception, mai chalunga always")

output:
    supriyo
    exception in code 'gender'
    I don't care , mai rookne nahi wala,bhar mai jay duniya and exception, mai chalunga always


-------Indexerror--------------

try:
    list1=[1,2,3,4]
    print(list1[10])
except IndexError as e:
    print("exception in code",e)
else:
    print("Hurray try block executed without any exception!!!!")
finally:
    print("I don't care , mai rookne nahi wala,bhar mai jay duniya and exception, mai chalunga always")

output:
exception in code list index out of range
I don't care , mai rookne nahi wala,bhar mai jay duniya and exception, mai chalunga always
```
